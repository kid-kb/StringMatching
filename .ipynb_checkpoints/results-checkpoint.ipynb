# String Matching Algorithms

### Brute force algorithm :

In this, given a pattern and a text, the algorithm searches for occurrences of the pattern within the text by checking each possible substring of the text that matches the pattern. The algorithm compares each substring of length equal to the pattern's length with the pattern itself and checks if they are equal.

The time complexity of the brute force algorithm for string matching is O(mn), where m is the length of the pattern and n is the length of the text. In the worst case, the algorithm compares each substring of the text with the pattern, which takes O(n) time, and it performs this operation for each character of the pattern, which takes O(m) time. Therefore, the total time complexity of the algorithm is O(mn).

### Knuth-Morris-Pratt (KMP) algorithm:

The KMP algorithm works by preprocessing the pattern string to create a prefix function. The prefix function is used to find the longest proper prefix of the pattern that is also a suffix of the pattern. This information is then used to skip comparisons of characters in the text that we know cannot match the pattern.

The time complexity of the KMP algorithm is O(n + m), where n is the length of the text and m is the length of the pattern. The preprocessing step takes O(m) time to compute the prefix function, and the matching step takes O(n) time to compare each character in the text with the pattern. However, the algorithm skips over comparisons of characters that we know cannot match the pattern, which makes it more efficient than the brute force algorithm.

### Boyer-Moore-Horspool (BMH) algorithm:

The BMH algorithm works by scanning the text string from left to right and comparing characters with the pattern string from right to left. If a mismatch occurs, the algorithm uses two heuristics to determine how many characters to skip in the text string before trying the next match. The first heuristic is the bad character rule, which uses the rightmost occurrence of a mismatched character in the pattern string to determine how many characters to skip in the text string. The second heuristic is the good suffix rule, which uses the longest suffix of the pattern that matches a prefix of the pattern to determine how many characters to skip in the text string.

The worst-case time complexity of the algorithm is O(nm).The time complexity of the BMH algorithm is O(n + m), where n is the length of the text and m is the length of the pattern.

### Rabin-Karp algorithm:

The Rabin-Karp algorithm works by using a rolling hash function to compute the hash value of the pattern and the hash value of all substrings of the text with the same length as the pattern. The algorithm then compares the hash values to determine if a match has occurred. If the hash values match, the algorithm performs a character-by-character comparison to confirm the match. If the hash values do not match, the algorithm moves on to the next substring.

The worst-case time complexity of the algorithm is O(nm), where n is the length of the text and m is the length of the pattern. However, the average case time complexity is O(n + m), which makes the algorithm efficient for most inputs. The performance of the algorithm depends on the choice of hash function and the size of the hash table used to store the hash values.

### Bitap algorithm, also known as the Baeza-Yates-Gonnet algorithm

The Bitap algorithm works by precomputing a set of bitmasks that represent the pattern. The bitmasks are then used to compare the pattern with the text in a series of bitwise operations. During the comparison, the algorithm keeps track of the number of errors and tries to minimize the number of errors by using a sliding window technique.

The time complexity of the Bitap algorithm is O(mn), where m is the length of the pattern and n is the length of the text. However, in practice, the algorithm is much faster than the brute force algorithm for most inputs.

The Bitap algorithm is a very useful algorithm for approximate string matching and is often used in practice. It is particularly useful for searching for patterns in noisy text, such as DNA sequences, where there may be errors or mutations. However, it can be sensitive to the choice of the error threshold and may not work well for patterns with a large number of errors.

```python
f = open('result_random.txt', 'r')
content = f.read()
f.close()
temp = content.split('\n')
time_random = []
for i in temp:
    temp2 = i.split(',')
    time_ran=[float(temp2[x]) for x in range(1,len(temp2)-1)]
    time_random.append(time_ran)
print(time_random)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# create data
x = [2,3,4,5,6,7,8,9,10]
# plot lines
plt.plot(x, time_random[0], label = "brute")
plt.plot(x, time_random[1], label = "kmp")
plt.plot(x, time_random[2], label = "bmh")
plt.plot(x, time_random[3], label = "rk")
plt.plot(x, time_random[4], label = "bitap")
plt.xlabel("String Length")
plt.ylabel("Time in Microseconds")
plt.legend()
plt.show()
```
