### Brute force algorithm :

Given a pattern and a text, the algorithm searches for occurrences of the pattern within the text. So, at every index i of the text, it checks whether the given pattern is present in the text starting from that particular index.

The time complexity of the brute force algorithm for string matching is O(mn), where m is the length of the pattern and n is the length of the text.

### Knuth-Morris-Pratt (KMP) algorithm:

The KMP algorithm works by preprocessing the pattern string to create a failure (prefix) function. The failure (prefix) function is used to find the longest proper prefix of the pattern that is also a suffix of the pattern. We can use this information to skip comparisons of characters in the text that we know cannot match the pattern, leading to faster performance.

The time complexity of the KMP algorithm is O(n + m), where n is the length of the text and m is the length of the pattern. The preprocessing step takes O(m) time to compute the failure function, and the matching step takes O(n) time to compare each character in the text with the pattern.

### Boyer-Moore-Horspool (BMH) algorithm:

The Boyer-Moore-Horspool (BMH) algorithm uses a preprocessing step to create a skip table. The skip table stores the maximum number of characters that can be skipped in the text when there is a mismatch between the current character in the text and the last character in the pattern. The BMH algorithm starts by comparing the last character of the pattern with the character at the current position in the text. If they match, it continues comparing the characters of the pattern with the characters of the text, moving from right to left. If there is a mismatch at some position, the skip table is used to determine how many characters to skip in the text before continuing the comparison.

The worst-case time complexity of the algorithm is O(nm + k), where n is the length of the text, m is the length of the pattern, k is the alphabet size.

### Rabin-Karp algorithm:

The Rabin-Karp algorithm works by using a rolling hash function to compute the hash value of the pattern and the hash value of all substrings of the text with the same length as the pattern. The algorithm then compares the hash values to determine if a match has occurred. If the hash values match, the algorithm performs a brute-force check. The choice of the hash function greatly affects the performance of the algorithm.

The worst-case time complexity of the algorithm is O(nm), where n is the length of the text and m is the length of the pattern. However, the average case time complexity is O(n + m).

### Bitap algorithm:

The Bitap (Shift-Or) algorithm is a bitwise algorithm for string matching. It works by precomputing a set of bitmasks that represent the pattern, and then uses a series of bitwise operations to compare the pattern with the text. During the comparison, the algorithm keeps track of the number of errors and tries to minimize it by using a sliding window technique.

The time complexity of the Bitap algorithm is O(mn), where m is the length of the pattern and n is the length of the text. More precisely the time complexity is O((n+k)[m/w]+m) (w is the word size and k is the alphabet size)

### Analysis of the above algorithms on Random strings

```python
f = open('result_random.txt', 'r')
content = f.read()
f.close()
temp = content.split('\n')
time_random = []
for i in temp:
    temp2 = i.split(',')
    time_ran=[float(temp2[x]) for x in range(1,len(temp2)-1)]
    time_random.append(time_ran)

print(time_random)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# create data
x = [2,3,4,5,6,7,8,9,10]
# plot lines
plt.plot(x, time_random[0], label = "brute")
plt.plot(x, time_random[1], label = "kmp")
plt.plot(x, time_random[2], label = "bmh")
plt.plot(x, time_random[3], label = "rk")
plt.plot(x, time_random[4], label = "bitap")
plt.xlabel("String Length")
plt.ylabel("Time in Microseconds")
plt.legend()
plt.show()
```

From the above graph, we can infer that the brute force algorithm performs the worst among all the above algorithms followed by rabin karp.

BMH performs the best and runs in a minimal time followed by the Bitap algorithm.

### Analysis of the above algorithms on English strings and words.

```python
import matplotlib.pyplot as plt
import numpy as np

# create data
x = [2,3,4,5,6,7,8,9,10]
# plot lines
plt.plot(x, time_random[0], label = "brute")
plt.plot(x, time_random[1], label = "kmp")
plt.plot(x, time_random[2], label = "bmh")
plt.plot(x, time_random[3], label = "rk")
plt.plot(x, time_random[4], label = "bitap")
plt.xlabel("String Length")
plt.ylabel("Time in Microseconds")
plt.legend()
plt.show()
```

From the above graph, we can infer that the brute force algorithm and the Rabin Karp algorithm almost have the same worst-case time complexity, and perform worse than the other algorithms.

BMH performs the best and runs the fastest.

## Conclusion

We can conclude, that Brute Force Algorithm is taking a long time since it has a worst-case time complexity of O(nm).

Rabin-Karp algorithm in these cases is performing worse possibly due to multiple hash collisions, which is causing the algorithm to do a character-by-character comparison for every substring, leading to a worst-case time complexity of O(nm).

BMH performs the best and runs the fastest since it uses a preprocessing step to compute a table of shifts(skip table) for each character in the alphabet, which can be used to quickly skip over characters in the text that cannot match the pattern, which can reduce the number of comparisons needed and lead to faster performance.

Bitap is performing well as it uses bitwise operations to perform fast comparisons between the pattern and the text.
